## Filter data

### Remove mean

```python
# raw_signal = e4_data["HR"]["afraid"][:,1]
raw_signal = eeg_data["afraid"][:,1]

mean = preprocessing.mean(raw_signal)

plt.figure()
plt.plot(range(len(raw_signal)),raw_signal, label = "raw_signal")
plt.title("Raw signal")
plt.hlines(mean,0,len(raw_signal),label="mean", color = "r")
plt.legend()

signal = preprocessing.remove_mean(raw_signal)

plt.figure()
plt.plot(range(len(signal)),signal, label = "raw_signal")
plt.title("Raw signal")
plt.legend()
```
